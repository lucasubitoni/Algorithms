```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
x = np.linspace(1,2,1000)
```

```python
i = 0
y_1 = np.zeros(len(x))
for time in x:
    y_1[i] = np.sin(2*np.pi*200*time) + np.sin(2*np.pi*400*time) 
    i+=1
    
plt.figure(figsize=(10,5))
plt.plot(y_1)
plt.show()
```

```python
i = 0
y_2 = np.zeros(len(x))
for value in x:
    if value<=1.5:
        y_2[i] = np.sin(2*np.pi*200*value)
    else:
        y_2[i] = np.sin(2*np.pi*400*value)
    i+=1
    
plt.figure(figsize=(10,5))
plt.plot(y_2)
plt.show()
```

```python
freqs_2, psd_2 = signal.welch(y_2, fs=1000)
freqs_1, psd_1 = signal.welch(y_2, fs=1000)

plt.figure(figsize=(10, 10))
plt.semilogx(freqs_1, psd_1)
plt.semilogx(freqs_2, psd_2)
plt.title('PSD: power spectral density')
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.tight_layout()
```
